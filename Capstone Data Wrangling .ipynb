```python
df1.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
# The table below in Heat Map-format shows the correlations between the variables. 
# Red conveys a positive correlation and Blue conveys a negative correlation.
# Some of the correlations are somewhat counterintuitive. 
# The most positive correlation for the Heat Rate (hr) is with the two natural gas price series: normally when the 
# gas price is high, only the most eficient plants will run. However, natural gas prices tend to be highest in the 
# Summer months because of the strong A/C load, when even inefficient plants run to help meet the huge demand spikes.
# This is precisely why multivariate regression analysis is needed, so all factors can be considered.
# We can also observe that a number of variables appear strongly positively or negatively correlated. 
# For this reason of so-called "multicollinearity", when variables are dropped the explanatory power is little reduced.
```
